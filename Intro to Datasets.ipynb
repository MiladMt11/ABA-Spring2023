## 6. Tax Dataset

```python
import os
Tax_files_path = 'D://DTU//DTU_Courses//Second Semester//Advanced BA//Project//ABA-Spring2023//Data//Tax Data//'
TX_df =  pd.DataFrame()

for files in os.listdir(Tax_files_path):
    
    year_tx_df = pd.read_csv(Tax_files_path+files,encoding='latin1')
    
    
    
    if files.endswith('.txt'):
        if files.upper().startswith('FY'):
            year = files[2:6]
        else:
            year = '20'+ files[:2]
            
            
        year_tx_df['Year'] = year
        year_tx_df.rename(columns=lambda x: ' '.join(x.split()),inplace=True)
    
        TX_df = pd.concat([TX_df, year_tx_df])
    
    else:
        continue
        
        
        
## 
        
for col in TX_df.columns:
    #print(col)
    col_cnt = TX_df[(TX_df[col].notna()) & (TX_df[col] != '') & (TX_df[col] != 'R') ][col].count()
    if col_cnt is not None:
        if col_cnt <5:
            TX_df.drop(columns=col,inplace=True)

TX_df.drop(columns=['Unnamed: 10','Unnamed: 14','Unnamed: 38','Unnamed: 56','US','DC'],inplace=True)


TX_df['item'].fillna(TX_df['ST'],inplace=True)
TX_df['item'].fillna(TX_df['Code'],inplace=True)
TX_df['item'].fillna(TX_df['Unnamed: 0'],inplace=True)


TX_df.drop(columns=['ST','Code','Unnamed: 0'],inplace=True)


TX_df_itemnames = pd.read_excel('D://DTU//DTU_Courses//Second Semester//Advanced BA//Project//ABA-Spring2023//Data//Tax Data//taxitemcodesanddescriptions.xls')
TX_df = TX_df.merge(TX_df_itemnames, left_on='item', right_on='Item Code')
```

```python
TX_df.head()
```

```python
TX_df.groupby(['Description']).size().reset_index().rename(columns={0:'count'})
```

```python
print('Min Year:',np.min(TX_df['Year']))
print('Max Year:',np.max(TX_df['Year']))
```

## 7. Population Dataset

```python
POP_df_raw = pd.read_csv('D://DTU//DTU_Courses//Second Semester//Advanced BA//Project//ABA-Spring2023//Data//nst-est2019-alldata.csv')

POPESTIMATE_cols = [col for col in list(POP_df_raw.columns) if col.startswith('POPESTIMATE') ]

POP_df = POP_df_raw.melt(id_vars=["NAME"], value_vars=POPESTIMATE_cols)


POP_df['variable'] = POP_df['variable'].apply(lambda x: x[11:])
POP_df.columns = ['State','Year','Population']
```
