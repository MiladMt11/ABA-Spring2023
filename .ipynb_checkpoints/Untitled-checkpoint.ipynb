```python
import numpy as np
import pandas as pd
import pickle
from sklearn.metrics import mean_squared_error
```

```python
path = './Data/Pickle Files/'
```

```python
forecast_df = pd.read_pickle(path + 'forecast_df.pickle')
forecast_df
```

```python
future_df = pd.read_pickle(path + 'future_data.pickle')
future_df
```

```python
df = pd.read_pickle(path + 'test_acc_df.pickle')
df
```

```python
can_acc_dict = {}
CAN_no = [10, 11, 5, 6, 7, 8, 9]

for i in CAN_no:
    can_acc_dict[f'CAN{i}_1'] = mean_squared_error(df[f'CAN{i}_1_true'], df[f'CAN{i}_1_pred'])
    
can_acc_df = pd.DataFrame(can_acc_dict, index=[0])
can_acc_df = can_acc_df.T.rename(columns={0:'Accuracy'})
can_acc_df
```
